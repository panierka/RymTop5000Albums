```python
df_g = df.copy()

def combine_genre_columns(row):
    primary = row['primary_genres']
    secondary = row['secondary_genres']
    seps = (',', ', ')

    def split_by_all(s: str):
        for sep in seps[1:]:
            s = s.replace(sep, seps[0])
        return s.split(seps[0])

    if primary == '':
        return split_by_all(secondary)

    if secondary == '':
        return split_by_all(primary)

    return split_by_all(primary) + split_by_all(secondary)

df_g.insert(loc=4, column='genres', value=df_g.apply(combine_genre_columns, axis=1))
df_g.drop(columns=['primary_genres', 'secondary_genres'], inplace=True)
df_g['genres'] = df_g['genres'].values.tolist()
genres = pd.DataFrame(df_g['genres'].explode())
genres.columns = ['genre']
genres.reset_index(inplace=True)
genres
```

```python
def find_genre(genre):
    mask = np.zeros(len(df)).astype(bool)
    indices = genres.loc[genres['genre'] == genre, 'index'].apply(lambda x: x - 1).tolist()
    mask[indices] = True
    return mask

preview_columns = ['release_name', 'artist_name', 'avg_rating', 'genres']

def show_genre(genre):
    return df_g.loc[find_genre(genre), preview_columns].sort_values('avg_rating', ascending=False)

show_genre('Neofolk')
```

```python
tmp = df_g.copy()
tmp = tmp.explode('genres')
tmp['release_decade'] = tmp['release_date'].map(lambda x: x.year // 10 * 10)
d_genre_counts = tmp.groupby(['release_decade', 'genres']).size()
d_genre_counts
```

```python
decades = d_genre_counts.index.get_level_values(0).unique().tolist()
decades
```

```python
from IPython.core.display_functions import display

for i, decade in enumerate(decades):
    tmp = d_genre_counts[decade].sort_values(ascending=False)
    total = tmp.sum()
    n = 10
    other = tmp.tail(-n).sum()
    tmp = tmp.head(n)
    tmp.loc['Pozostałe'] = other
    tmp.plot.pie(title=decade, ylabel='', autopct='%1.1f%%', pctdistance=0.9)
    plt.show()
    top_genre = tmp.head(1).index[0]
    display(
        df_g.loc[
            find_genre(top_genre)
            & (df_g['release_date'].map(lambda x: x.year // 10 * 10) == decade),
            ['artist_name', 'release_name', 'avg_rating', 'release_date']
        ].sort_values('avg_rating', ascending=False).head(3)
    )
```

```python
def show_genres_popularity_rise(genre_list):
    tmp = df_g.copy()
    tmp = tmp.explode('genres')
    tmp['release_year'] = tmp['release_date'].map(lambda x: x.year)

    genre_popularity_in_time = pd.DataFrame(tmp.groupby(['genres', 'release_year']).size())
    genre_popularity_in_time.columns = ['count']
    genre_popularity_in_time = genre_popularity_in_time.loc[genre_list].unstack(0)
    genre_popularity_in_time.fillna(0, inplace=True)
    genre_popularity_in_time.columns = genre_popularity_in_time.columns.get_level_values(1)
    genre_popularity_in_time.index = genre_popularity_in_time.index.sort_values()
    genre_popularity_in_time = genre_popularity_in_time.astype(int)
    genre_popularity_in_time = genre_popularity_in_time.cumsum()
    genre_popularity_in_time.plot.line(legend=True, figsize=(20, 10), style='o-')
    plt.show()
```

```python
def get_top_n_genres(n):
    return genres_pop_rank.head(n)['genre'].tolist()
```

```python
top10_genres = get_top_n_genres(10)
show_genres_popularity_rise(top10_genres)
```
